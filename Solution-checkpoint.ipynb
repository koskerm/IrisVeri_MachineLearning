```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

encoded = le.fit_transform(labeled['iris'])
```

```python
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

#np.random.seed(42)

X = labeled.drop('iris', axis=1)
y = labeled['iris']


#train_pct_index = int(0.8 * len(X))
#X_train, X_test = X[:train_pct_index], X[train_pct_index:]
#y_train, y_test = y[:train_pct_index], y[train_pct_index:]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
X_train, y_train
```

```python
from sklearn.metrics import accuracy_score, log_loss
classifiers = [
    LinearSVC(),
    KNeighborsClassifier(),
    LogisticRegression(),
    RandomForestClassifier(),               
                  ]
 
# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)
 
for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Sonuc****')
    train_predictions = clf.predict(X_test)
    acc = accuracy_score(y_test, train_predictions)
    print("Dogruluk: {:.2%}".format(acc))
    
    log_entry = pd.DataFrame([[name, acc*100, 11]], columns=log_cols)
    log = log.append(log_entry)
    
    print("="*30)
```

```python
log
```

```python
lin = LinearSVC()
kn = KNeighborsClassifier()
logr = LogisticRegression()
random_f = RandomForestClassifier()

models = {'LinearSVC': lin.fit(X_train, y_train),
          'KNeighborsClassifier': kn.fit(X_train, y_train),
          'LogisticRegression': logr.fit(X_train, y_train),
          'RandomForestClassifier': random_f.fit(X_train, y_train)}

for name, value in models.items():
    print(f'{name} --> {value.score(X_test, y_test)}')
```

```python
# KNeighbours scoring

from sklearn.model_selection import cross_validate

def cross_validation(model, X, y, cv=5):
    scoring = ['accuracy', 'precision', 'recall', 'f1']
    results = cross_validate(estimator=model,
                           X=X,
                           y=y,
                           cv=cv,
                           scoring=scoring,
                           return_train_score=True)
    
    return results

cross_validation(kn, X, y)
```

```python
X = labeled2[["sepal length", "sepal width"]]
y = labeled2["iris"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.inspection import DecisionBoundaryDisplay

n_neighbors = 5

# import some data to play with
data = labeled2

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = data[["sepal length", "sepal width"]]
y = data["iris"]

# Create color maps
cmap_light = ListedColormap(["orange", "cyan", "cornflowerblue"])
cmap_bold = ["darkorange", "c", "darkblue"]

for weights in ["uniform", "distance"]:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    _, ax = plt.subplots()
    DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        cmap=cmap_light,
        ax=ax,
        response_method="predict",
        plot_method="pcolormesh",
        xlabel='Sepal Length',  # sepal length stununu x_label olarak kullanyoruz
        ylabel='Sepal Width',  # sepal width stununu y_label olarak kullanyoruz
        shading="auto",
    )

    # Plot also the training points
    sns.scatterplot(
        x=X.to_numpy()[:, 0],  # sepal length stununu x ekseninde kullanyoruz
        y=X.to_numpy()[:, 1],  # sepal width stununu y ekseninde kullanyoruz
        hue=y,  # iris snflarn renklerle gsteriyoruz
        palette=cmap_bold,
        alpha=1.0,
        edgecolor="black",
    )
    plt.title(
        "KNN gorsellestirme (k = %i, agirlik = '%s')" % (n_neighbors, weights)
    )

plt.show()
```
